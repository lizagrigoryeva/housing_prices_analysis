## Какие задачи можно сформулировать и как классификацию, и как регрессию?

Главное правило: всё зависит от того, **что именно мы предсказываем** (целевая переменная), а не от самой бизнес-задачи. Вот как можно переформулировать задачи из таблицы:

### Задачи, которые подходят под оба варианта:

**2. Вернёт ли клиент кредит**
- Классификация: вернёт / не вернёт (0 или 1)
- Регрессия: сколько денег потеряет банк (сумма ущерба) или вероятность дефолта от 0 до 1

**3. Когда пациенту нужно принять лекарство**
- Классификация: временной интервал — утро/день/вечер или «срочно/потом/не нужно»
- Регрессия: точное время в часах до приёма или временная метка

**5. Выбор сегмента клиентов для промо**
- Классификация: откроет письмо или нет (да/нет), либо отнесение к группе A/B/C
- Регрессия: ожидаемый open rate (в процентах) или прогнозируемый прирост прибыли

**6. Поиск брака на производстве (по фото)**
- Классификация: бракованный / нормальный, или тип дефекта (царапина/трещина/...)
- Регрессия: оценка серьёзности дефекта (например, от 0 до 10) или вероятность того, что деталь бракованная

**7. Как разместить товары на полке**
- Классификация: выбор типа выкладки — план А / Б / В
- Регрессия: прогноз продаж в штуках или ожидаемый рост выручки для конкретной раскладки

**8. Поиск сайтов по текстовому запросу**
- Классификация: релевантен документ запросу или нет (для каждого результата)
- Регрессия: оценка релевантности (скор от 0 до 1), прогноз CTR или ожидаемая позиция в выдаче

**10. Детекция аномалий в трафике**
- Классификация: аномалия / норма
- Регрессия: «степень аномальности» (числовая оценка отклонения) или прогноз нормального уровня трафика

---


**1. Предсказание цены дома** — естественная регрессия.

**4. Выбор лекарства из нескольких** — по сути мультиклассовая классификация (какое из N лекарств лучше).







---

### Кластеризация (Clustering)
*Группируем объекты по схожести, без заранее известных ответов.*

**9. Сегментация клиентов** 
Кластеризуем клиентов по истории покупок и демографии, чтобы найти группы со схожим поведением (например, «экономные», «импульсивные», «лояльные»).

**2. Клиенты по кредитам**  
Можно сначала разбить клиентов на кластеры по профилю риска (доход, кредитная история, возраст), а потом уже внутри кластеров строить модели дефолта.

**5. Промо-сегменты**  
Кластеризация клиентов по реакции на прошлые акции помогает выделить группы, которым стоит слать разные предложения.

**5. Поиск дефектов**  
Группируем по виду дефектов

**7. Размещение товаров**  
Кластеризуем товары по паттернам покупок («часто берут вместе») или по предпочтениям покупателей — это помогает логичнее расставлять товары на полках.

**10. Аномалии в трафике**  
Сначала строим кластеры «нормального» трафика, а запросы, которые не вписываются ни в один кластер, помечаем как подозрительные.

---

### Ассоциативные правила (Association)
*Ищем закономерности вида «если А, то часто и Б».*

**7. Размещение товаров на полке**  
Классический market basket analysis: «покупатели, берущие пасту, часто берут и соус» => ставим их рядом.

**5. Промо-коммуникации**  
Ассоциации между товарами в промо и последующими покупками: «после скидки на кофе растёт продажа кружек» => учитываем при планировании акций.

**8. Поиск по сайту**  
«Пользователи, искавшие Х, часто кликали на статью У» => можно улучшать выдачу или предлагать похожие запросы.

**2. Кредитный скоринг**  
Поиск паттернов в кредитных историях: например, «клиенты с тремя микрозаймами за месяц и просрочкой по коммуналке чаще не возвращают кредит».

---

### Уменьшение размерности (Dimensionality Reduction)
*Сжимаем признаки, чтобы упростить модель или визуализировать данные.*

**6. Детекция брака по фото**  
Изображения — это тысячи пикселей. Можно применить PCA или автоэнкодеры, чтобы получить компактное представление фото перед классификацией.

**8. Текстовый поиск**  
Текст => векторы большой размерности. Методы помогают сжать представление запросов и документов для быстрого поиска.

**9. Сегментация клиентов**  
Перед кластеризацией можно уменьшить число признаков (доход, частота покупок, средний чек и т.д.) через PCA — это уберёт шум и ускорит работу алгоритмов.

**10. Аномалии в трафике**  
Сжимаем признаки запросов (время, источник, тип устройства) до 2–3 компонент — аномалии часто лучше видны в низкоразмерном пространстве.

**1. Предсказание цены дома**  
Если признаков много (площадь, район, этаж, год постройки, инфраструктура...), PCA помогает убрать мультиколлинеарность и упростить модель регрессии.

- Классификация

- Классификация, кластеризация (группировка одинакового вида писем)

5) Прогнозирование трафика в картах (яндекс карты и др.)

# 5. Training

```python
result_MAE = pd.DataFrame(columns=['model', 'train', 'test'])
result_RMSE = pd.DataFrame(columns=['model', 'train', 'test'])
```

## Linear Regression

```python
model_linreg = LinearRegression()
```

```python
X_train = X_train.drop(columns=['latitude', 'longitude', 'listing_id'])
X_test = X_test.drop(columns=['latitude', 'longitude', 'listing_id'])
y_train = train_df['price']
y_test = test_df['price']
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
model_linreg.fit(X_train, y_train)
pred_linreg_test = model_linreg.predict(X_test)
test_df['price_linreg'] = pred_linreg_test

pred_linreg_train = model_linreg.predict(X_train)
train_df['price_linreg'] = pred_linreg_train
```

```python
MAE_linreg_test = mean_absolute_error(y_test, pred_linreg_test)
MAE_linreg_train = mean_absolute_error(y_train, pred_linreg_train)

RMSE_linreg_test = np.sqrt(mean_squared_error(y_test, pred_linreg_test))
RMSE_linreg_train = np.sqrt(mean_squared_error(y_train, pred_linreg_train))
```

```python
result_MAE.loc[0] = {'model': 'linear_regression',
                     'train': MAE_linreg_train,
                     'test': MAE_linreg_test}

result_RMSE.loc[0] = {'model': 'linear_regression',
                     'train': RMSE_linreg_train,
                     'test': RMSE_linreg_test}
```

## Decision Tree

```python
model_dec_tree = DecisionTreeRegressor(random_state=21)
```

```python
model_dec_tree.fit(X_train, y_train)
pred_dec_tree_test = model_dec_tree.predict(X_test)
test_df['price_dec_tree'] = pred_dec_tree_test

pred_dec_tree_train = model_dec_tree.predict(X_train)
train_df['price_dec_tree'] = pred_dec_tree_train
```

```python
MAE_dec_tree_test = mean_absolute_error(y_test, pred_dec_tree_test)
MAE_dec_tree_train = mean_absolute_error(y_train, pred_dec_tree_train)

RMSE_dec_tree_test = np.sqrt(mean_squared_error(y_test, pred_dec_tree_test))
RMSE_dec_tree_train = np.sqrt(mean_squared_error(y_train, pred_dec_tree_train))
```

```python
result_MAE.loc[1] = {'model': 'decision_tree',
                     'train': MAE_dec_tree_train,
                     'test': MAE_dec_tree_test}

result_RMSE.loc[1] = {'model': 'decision_tree',
                     'train': RMSE_dec_tree_train,
                     'test': RMSE_dec_tree_test}
```

```python
MAE_naive_median_test = mean_absolute_error(y_test, test_df['median_price'])
MAE_naive_median_train = mean_absolute_error(y_train, train_df['median_price'])

RMSE_naive_median_test = np.sqrt(mean_squared_error(y_test, test_df['median_price']))
RMSE_naive_median_train = np.sqrt(mean_squared_error(y_train, train_df['median_price']))
```

MAE показывает средний модуль разности между фактическими (истинными) значениями и значениями, предсказанными моделью.

```python
result_MAE
```

RMSE показывает средний размер ошибки прогноза модели в тех же единицах, что и целевая переменная. Чем меньше значение RMSE, тем точнее модель предсказывает данные; значение 0 означает идеальное соответствие.
